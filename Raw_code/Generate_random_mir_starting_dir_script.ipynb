```python
def get_gridsize():
    gridsize = []
    for i in range(2):
        gridsize.append(random.randint(0, 1000))
    
    return gridsize
```

```python
def get_location():
    starting = []
    for j in range(2):
        starting.append(random.randint(0, min(gridsize)))
    
    dir_map = {0:'N', 1:'S', 2:'W', 3:'E'}
    ram = random.randint(0,3)
    starting.append(dir_map[ram])

    return starting
```

```python
def get_mir():
    all_mir = []
    for z in range(0, random.randint(0, min(gridsize))):
        mir = []
        mir_map = {0:'\\', 1:'/'}
        for k in range(2):
            mir.append(random.randint(0, min(gridsize)))

        mir_ram = random.randint(0,1)
        mir.append(mir_map[mir_ram])
        all_mir.append(mir)
    return all_mir
```
